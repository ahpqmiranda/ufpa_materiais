# eixos

Os eixos são elementos mecânicos / estruturais responsáveis pela transmissão de energia cinética entre dois ou mais conjuntos mecânicos

os eixos podem incluir engrenagens, polias, correias, acoplamentos, etc. que transmitem a energia cinética disponível

Os eixos são montados em __mancais_ podendo apresentar as configurações:
    1. montagem de sela
    2. balanço
    3. saliente
    
As cargas identificadas em eixos são:
    1. cargas de flexão
    2. cargas de torção
    3. cargas axiais
    
As cargas de flexão e torção são as mais comuns e são as que serão estudadas neste curso, as cargas axiais são mais comuns em eixos de transmissão de potência hidráulica e pneumática. As cargas de torção são devido ao torque transmitido, enquanto as cargas de flexão são originárias de cargas externas, como peso, forças de acoplamento de
 engrenagens, polias, etc.
 
## Natureza dos carregamentos em eixos

A depender da aplicação, os carregamentos nos eixos podem ser constantes, ou variar com o tempo, podendo ainda serem ciclicos, com torques alternados, repetitivos ou pulsantes.

## Tensões em eixos
As tensões flutuantes devidos a flexão e torção, e carregamentos alternadas podem ser estimados da seguinte forma:

Ma = Momento fletor (Nm)
c = braço de alavanca (m)
I = Momento de inércia (m⁴)
Kf = concentrador de tensão para fadiga por flexão
Kfs = concentrador de tensão para fadiga por torção



```python
M_a, T_a, c, d, I, J, k_f, k_fs, sigma_a, tau_a = sy.symbols('M_a T_a c d I J k_f k_fs sigma_a tau_a', real=True)

tensao_flex = sy.Eq(sigma_a, k_f * M_a * c / I)
tensao_flex
```

```python
tensao_tor = sy.Eq(tau_a, k_fs * M_a * c / J)
tensao_tor
```

## carregamentos em eixos circulares, temos:

```python
from sympy import pi
tensao_flex = sy.Eq(tau_a, k_f * 32 * M_a / (pi * d ** 3))
tensao_flex
```

```python
tensao_tor = sy.Eq(tau_a, k_fs * 16 * T_a / (pi * d ** 3))
tensao_tor
```

Dadas as equações para carregamentos por torção e flexão nos eixos, é possível estimar, através das teorias de falha, os valores de falha e fatores de segurança para determinado eixo.
