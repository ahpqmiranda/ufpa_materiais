### Dados necess√°rios:

```python
massa_escada = 195 # [lbs]
coeficiente_atrito = 0.35 # [-]
g = sy.physics.units.g # [m/s^2]
g.convert_to(sy.physics.units.ft/sy.physics.units.s**2)
Lx = 3 # [ft]
Ly = 10 # [ft]
```

Elementos a serem calculados:

```python
normal_B, normal_A, Fat_A, theta = sy.symbols('normal_B normal_A Fat_A theta')
x = 10 * sy.cos(theta) - 3
y = 10 * sy.sin(theta)
```

### Calculando: <br>
somat√≥rio das for√ßas em x:

```python
Fat_A = coeficiente_atrito * normal_A
somatorio_fx = sy.Eq(normal_B - Fat_A, 0)
somatorio_fx
```

Somat√≥rio das for√ßas em Y:

```python
somatorio_fy = sy.Eq( - massa_escada * g + normal_A, 0)
somatorio_fy
```

rela√ß√£o entre as for√ßas em x e y:

```python
solucao = sy.solve([somatorio_fx, somatorio_fy], [normal_B, normal_A, Fat_A])
solucao
```

Somat√≥rio dos momentos em rela√ß√£o ao ponto A:

```python
somatorio_Ma = sy.Eq(- normal_A * x + normal_B * y, 0)
somatorio_Ma
```

Combina√ß√£o das equa√ß√µes: <br> valores das normais (solucao) na eq dos momentos (somatorio_Ma):

```python
eq_final = sy.simplify(somatorio_Ma.subs(solucao))
eq_final
```

a equa√ß√£o na linha anterior descreve a rela√ß√£o entre as medidas de for√ßas em x, y e os momentos em rela√ß√£o ao ponto A para cada √¢ngulo, o que nos d√° a solu√ß√£o final em radianos:

```python
sy.solve(eq_final, theta, dict=True)
```

salvando a solu√ß√£o em uma lista:

```python
theta_valores = sy.solve(eq_final, theta)
```

Isolar as solu√ß√µes da equa√ß√£o com a fun√ß√£o solve e salvar no vetor_solucao:

```python
theta_graus = []
for angle in theta_valores:
    theta_graus.append(m.degrees(angle))
data1 = pd.DataFrame({
    'theta em rad': theta_valores,
    'theta em graus': theta_graus
})
data1
```
