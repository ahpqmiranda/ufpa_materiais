### Dados dispon√≠veis

```python
Lx = 0.2 # [m]
Ly = 0.1 # [m]
y = (50 * x) ** 0.5
rho = 10e6 # [kg/m¬≥]
pi = m.pi
```

Paraboloide de Revolu√ß√£o: qual o raio de gira√ß√£o?
notas:
o raio de revolu√ß√£o varia com y em 1:1
0 < x < 200
0 < y < 100
0 < sqrt(50x) <100
o c√°lculo de momento de in√©rcia ao longo do eixo x, √©:
Ix = integral(y¬≤)dm
Ix - integral(y¬≤) rho * PI * r dr dx

```python
inercia_x = sy.integrate((y ** 2) * pi * r, (r, 0, y), (x, 0, Lx))
print('momento de in√©rcia em X: {} m^4'.format(inercia_x))
```

O raio de gira√ß√£o √© dado por k = sqrt(Ix / m), logo, devemos calcular a massa do s√≥lido de reolu√ß√£o:
o volume do s√≥lido √© dado pela integral de revolu√ß√£o:
integral(PI * (y ** 2)) dx

```python
massa_solido = sy.integrate(pi * rho * r , (r, 0, y), (x, 0, Lx))
print('A massa do s√≥lido de revolu√ß√£o:{:.3f} kg'.format(massa_solido))
```

O raio de gira√ß√£o ser√°:

```python
k = (inercia_x / massa_solido) ** 0.5
scientific = "{:e} m".format(k)
scientific
```
