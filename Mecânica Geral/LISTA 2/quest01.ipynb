## Dados Utilizados

```python
v = 0.28            # [m]
rho_rejeito = 1800  # [kg/m]
m = v * rho_rejeito # [kg]
g = 9.81            # [m/s]
D_xa = 1.9           # [m]
D_ya = 0.4           # [m]
DF_x = 0.3           # [m]
E_xa = 0.1           # [m]
E_ya = 0.25          # [m]
```

### Equacionamentos:
questo 1.A

somatrio de foras em x:
consideraes: c_x = d_x e c_y = d_y (mesmo elemento estrutural)
(l-se f_x: "fora no ponto f na direo x" e d_y: "fora no ponto d na direo y"

```python
sum_fx = sp.Eq(c_x + f_x, 0)
sum_fx
```

```python
sp.solve(sum_fx)
```

somatrio de foras em y:

```python
sum_fy = sp.Eq(c_y + f_y - m * g, 0)
sum_fy
```

Somatrio de momentos no ponto d:

```python
sum_Md = sp.Eq( - m * g * D_xa - c_y * D_ya + f_y * DF_x, 0)
sum_Md
```

```python
sum_Md.subs(sum_fy)
```

```python

sp.solve(sum_Md, f_y)
```
