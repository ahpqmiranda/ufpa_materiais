```python
from sympy import symbols, Eq, solve
from IPython.display import Image
```

```python
Image(filename="12-40.png")
```

```python
# Definindo variveis
vB, vA = symbols('vB vA')  # Velocidades nos pontos B e no bloco A
```

```python
# Relao entre velocidades devido  configurao do sistema de polias
# A velocidade de B  duas vezes a velocidade de A (porque a corda se divide em duas partes mveis ligadas ao bloco A)
relation = Eq(vB, 2 * vA)
relation
```

```python
# Substituindo o valor de vB = 6 m/s para encontrar vA
vA_value = solve(relation.subs(vB, 6), vA)[0]

vA_value
```
