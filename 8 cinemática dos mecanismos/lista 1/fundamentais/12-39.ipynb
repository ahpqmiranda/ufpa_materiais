```python
from sympy import symbols, Eq, solve
from IPython.display import Image
```

```python
# Definindo variáveis
vA, vD = symbols('vA vD')  # Velocidade no ponto A e no bloco D
```

```python
# Relação entre velocidades devido à configuração do sistema de polias
# A velocidade de A é duas vezes a velocidade de D (porque a corda se divide em duas partes móveis ligadas a D)
relation = Eq(vA, 2 * vD)
relation
```

```python
# Substituindo o valor de vA = 3 m/s para encontrar vD
vD_value = solve(relation.subs(vA, 3), vD)[0]

vD_value
```
