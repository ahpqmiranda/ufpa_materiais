```python
from sympy import symbols, diff, sqrt
from IPython.display import Image
```

```python

Image(filename="12-34.png")
```

```python
# Definindo variveis
t = symbols('t')  # Tempo
r = 0.1 * t**3  # Posio radial em funo do tempo
theta = 4 * t**(3/2)  # Posio angular em funo do tempo
theta
```

```python
# Derivadas para calcular velocidade e acelerao
dr_dt = diff(r, t)  # Velocidade radial
dtheta_dt = diff(theta, t)  # Velocidade angular
dtheta_dt
```

```python
# Velocidade tangencial
v_tangential = r * dtheta_dt
v_tangential
```

```python
# Intensidade da velocidade totalm
v_total = sqrt(dr_dt**2 + v_tangential**2)
v_total
```

```python
# Aceleraes
a_radial = r * dtheta_dt**2  # Acelerao centrpeta
a_tangential = diff(v_tangential, t)  # Acelerao tangencial
a_total = sqrt(a_radial**2 + a_tangential**2)
a_total
```

```python
# Substituindo t = 1.5 s para obter valores numricos
t_value = 1.5
v_total_value = v_total.subs(t, t_value)
a_total_value = a_total.subs(t, t_value)

v_total_value, a_total_value
print(f'velocidade total para t=1.5s : {v_total_value:.3f} m/s \nAcelerao : {a_total_value:.3f} m/s')
```
