```python
from sympy import symbols, diff, sqrt
from IPython.display import Image
```

```python

Image(filename="12-34.png")
```

```python
# Definindo variáveis
t = symbols('t')  # Tempo
r = 0.1 * t**3  # Posição radial em função do tempo
theta = 4 * t**(3/2)  # Posição angular em função do tempo
theta
```

```python
# Derivadas para calcular velocidade e aceleração
dr_dt = diff(r, t)  # Velocidade radial
dtheta_dt = diff(theta, t)  # Velocidade angular
dtheta_dt
```

```python
# Velocidade tangencial
v_tangential = r * dtheta_dt
v_tangential
```

```python
# Intensidade da velocidade totalm
v_total = sqrt(dr_dt**2 + v_tangential**2)
v_total
```

```python
# Acelerações
a_radial = r * dtheta_dt**2  # Aceleração centrípeta
a_tangential = diff(v_tangential, t)  # Aceleração tangencial
a_total = sqrt(a_radial**2 + a_tangential**2)
a_total
```

```python
# Substituindo t = 1.5 s para obter valores numéricos
t_value = 1.5
v_total_value = v_total.subs(t, t_value)
a_total_value = a_total.subs(t, t_value)

v_total_value, a_total_value
print(f'velocidade total para t=1.5s é: {v_total_value:.3f} m/s \nAceleração é: {a_total_value:.3f} m/s²')
```
