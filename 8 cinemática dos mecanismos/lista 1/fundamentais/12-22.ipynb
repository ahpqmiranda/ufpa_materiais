```python
from IPython.display import Image
from sympy import symbols, sin, cos, solve, Eq, sqrt, pi
```

```python
# Definindo variáveis e constantes
t, g = symbols('t g')  # Tempo e gravidade
vA, angle = symbols('vA angle')  # Velocidade inicial e ângulo de lançamento
x, y = symbols('x y')  # Coordenadas
```

```python

# Definindo componentes da velocidade inicial
vAx = vA * cos(angle)  # Componente horizontal
vAy = vA * sin(angle)  # Componente vertical
```

```python
# Equações de movimento
x_eq = vAx * t  # Movimento horizontal
y_eq = vAy * t - (1/2) * g * t**2  # Movimento vertical
```

```python
# Determinando o tempo de voo (quando y = 0)
t_total = solve(Eq(y_eq, 0), t)[1]  # Pegando a solução positiva
t_total
```

```python
# Determinando o alcance (R)
R = x_eq.subs(t, t_total)
R
```

```python
# Determinando a velocidade escalar no impacto
vx_final = vAx  # Componente horizontal permanece constante
vy_final = vAy - g * t_total  # Componente vertical no impacto
v_final = sqrt(vx_final**2 + vy_final**2)  # Velocidade escalar
v_final
```

```python
# Substituindo valores numéricos: vA = 10 m/s, angle = 30 graus, g = 9.81 m/s²
numerical_values = {vA: 10, angle: 30 * pi / 180, g: 9.81}
R_value = R.subs(numerical_values)
v_final_value = v_final.subs(numerical_values)
R_value, v_final_value
```
