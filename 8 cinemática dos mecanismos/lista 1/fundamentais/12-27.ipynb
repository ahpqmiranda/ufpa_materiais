```python
from sympy import symbols, diff, sqrt
from IPython.display import Image
```

```python
Image(filename="12-27.png")
```

```python
# Definindo variáveis
t = symbols('t')  # Tempo
v = 0.0625 * t**2  # Velocidade escalar em m/s
r = 40  # Raio da trajetória circular em metros
```

```python
# Calculando a aceleração tangencial (derivada da velocidade em relação ao tempo)
at = diff(v, t)
at
```

```python
# Calculando a aceleração centrípeta (v^2 / r)
ac = v**2 / r
ac
```

```python
# Aceleração total (resultante): sqrt(at^2 + ac^2)
a_total = sqrt(at**2 + ac**2)
a_total
```

```python
# Substituindo t = 10 s para calcular os valores numéricos
t_value = 10
at_value = at.subs(t, t_value)
ac_value = ac.subs(t, t_value)
a_total_value = a_total.subs(t, t_value)

at_value, ac_value, a_total_value
```
