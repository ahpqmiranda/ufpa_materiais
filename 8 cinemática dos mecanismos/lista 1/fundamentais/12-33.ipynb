```python
# Definindo variáveis
v, r = symbols('v r')  # Velocidade escalar e raio
theta_dot = v / r  # Velocidade angular
theta_dot
```

```python
# Substituindo os valores numéricos: v = 16.5 m/s, r = 120 m
theta_dot_value = theta_dot.subs({v: 16.5, r: 120})

print(f'Velocidade angular: {theta_dot_value:.3f} rad/s')
```
