```python
from sympy import symbols, cos, sin, sqrt, pi
from IPython.display import Image
```

```python
# Definindo variveis e constantes
t = symbols('t')  # Tempo
vA, vB = 15, 10  # Velocidades escalares dos barcos (m/s)
theta = 30 * pi / 180  # ngulo em radianos
```

```python
# Posies dos barcos
xA = vA * t * cos(theta)  # Componente x do barco A
yA = vA * t * sin(theta)  # Componente y do barco A
xB = 0  # Barco B se move apenas na direo y
yB = vB * t
```

```python
# Distncia entre os barcos
distance = sqrt((xA - xB)**2 + (yA - yB)**2)
distance
```

```python
# Substituir t = 4 s para calcular a distncia numrica
distance_value = distance.subs(t, 4)

distance_value
```
