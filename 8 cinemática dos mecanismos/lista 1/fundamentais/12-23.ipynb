```python

# Define symbols
v_A = symbols('v_A')  # Initial velocity to be determined
theta = 30 * pi / 180  # Angle in radians
g = 9.81  # Gravity (m/s^2)
x = 10  # Horizontal distance (m)
y_0 = 1.5  # Initial height (m)
y_f = 3  # Final height (m)

# Horizontal motion equation
t = symbols('t')
x_eq = Eq(x, v_A * cos(theta) * t)

# Solve for time (t)
t_expr = solve(x_eq, t)[0]

# Vertical motion equation
y_eq = Eq(y_f, y_0 + v_A * sin(theta) * t_expr - (1/2) * g * t_expr**2)

# Solve for initial velocity (v_A)
v_A_value = solve(y_eq, v_A)
v_A_value
```
