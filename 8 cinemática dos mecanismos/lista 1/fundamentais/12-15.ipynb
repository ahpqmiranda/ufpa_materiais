```python
from sympy import symbols, integrate, Eq, solve
from IPython.display import Image
```

```python
Image('12-15.png', width=200)
```

```python
# Definindo as variveis
t = symbols('t')  # Tempo
x, y = symbols('x y')  # Coordenadas

# Componentes de velocidade
vx = 32 * t  # Velocidade na direo x
vy = 8       # Velocidade na direo y

# Integrao para determinar as posies em funo do tempo
x_t = integrate(vx, t)  # Posio x(t)
y_t = integrate(vy, t)  # Posio y(t)

# Adicionando condies iniciais: x(0) = 0, y(0) = 0
x_t = x_t + symbols('C1')  # Constante de integrao para x
y_t = y_t + symbols('C2')  # Constante de integrao para y
```

```python
# Determinando as constantes de integrao
C1 = solve(Eq(x_t.subs(t, 0), 0), symbols('C1'))[0]
C2 = solve(Eq(y_t.subs(t, 0), 0), symbols('C2'))[0]
C1, C2
```

```python
# Substituindo as constantes nas equaes
x_t = x_t.subs(symbols('C1'), C1)
y_t = y_t.subs(symbols('C2'), C2)
x_t, y_t
```

```python
# Eliminar t para determinar y em funo de x
t_in_terms_of_x = solve(Eq(x_t, x), t)[0]  # Resolver t em termos de x
y_in_terms_of_x = y_t.subs(t, t_in_terms_of_x)  # Substituir t na equao de y
t_in_terms_of_x, y_in_terms_of_x
```
