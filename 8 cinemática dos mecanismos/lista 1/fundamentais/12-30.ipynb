```python
from sympy import symbols, diff, sqrt, atan, pi
from IPython.display import Image
```

```python
Image(filename="12-30.png")
```

```python
# Definindo variveis
x = symbols('x')  # Posio x
v_scalar = 6  # Velocidade escalar (m/s)
at = 1.8  # Acelerao tangencial (m/s)
```

```python
# Equao da trajetria
y = (1/8) * x**2
y
```

```python
# Derivada de y em relao a x para encontrar a inclinao
dy_dx = diff(y, x)
theta = atan(dy_dx)  # ngulo da inclinao
theta
```

```python
# Calculando a acelerao normal (centrpeta)
an = (v_scalar**2) * abs(dy_dx) / sqrt(1 + dy_dx**2)  # Frmula para acelerao normal
an
```

```python
# Acelerao total
a_total = sqrt(at**2 + an**2)
a_total
```

```python
# Substituir x = 3 m
x_value = 3
theta_value = theta.subs(x, x_value) * 180 / pi  # Converter para graus
an_value = an.subs(x, x_value)
a_total_value = a_total.subs(x, x_value)

theta_value, an_value, a_total_value
```
