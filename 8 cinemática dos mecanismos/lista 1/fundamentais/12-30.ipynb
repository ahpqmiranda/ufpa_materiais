```python
from sympy import symbols, diff, sqrt, atan, pi
from IPython.display import Image
```

```python
Image(filename="12-30.png")
```

```python
# Definindo variáveis
x = symbols('x')  # Posição x
v_scalar = 6  # Velocidade escalar (m/s)
at = 1.8  # Aceleração tangencial (m/s²)
```

```python
# Equação da trajetória
y = (1/8) * x**2
y
```

```python
# Derivada de y em relação a x para encontrar a inclinação
dy_dx = diff(y, x)
theta = atan(dy_dx)  # Ângulo da inclinação
theta
```

```python
# Calculando a aceleração normal (centrípeta)
an = (v_scalar**2) * abs(dy_dx) / sqrt(1 + dy_dx**2)  # Fórmula para aceleração normal
an
```

```python
# Aceleração total
a_total = sqrt(at**2 + an**2)
a_total
```

```python
# Substituir x = 3 m
x_value = 3
theta_value = theta.subs(x, x_value) * 180 / pi  # Converter para graus
an_value = an.subs(x, x_value)
a_total_value = a_total.subs(x, x_value)

theta_value, an_value, a_total_value
```
