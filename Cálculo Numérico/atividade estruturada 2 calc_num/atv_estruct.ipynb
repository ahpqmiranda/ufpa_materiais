### Questo 1: converter o sistema em matriz triangular superior:
modo 1:

```python
def convert_matrix(matrix):
    matrix_triangular = np.zeros((len(matrix), len(matrix)))
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if i <= j:
                matrix_triangular[i][j] = matrix[i][j]
    return matrix_triangular
```

```python
matrix = np.array([[2, 2, -1, 6], [1, 4, -1, 4], [1, -1, 4, -1]])
matrix
```

```python
matrix_triangular = convert_matrix(matrix)
matrix_triangular
```

# Questo 2: Utilizar a triangulao de Gauss para resolver o sistema:

```python
def gauss_triangular(matrix):
    matrix_triangular = convert_matrix(matrix)
    for i in range(len(matrix)):
        for j in range(i+1, len(matrix)):
            if matrix_triangular[i][i] == 0:
                print('No  possvel resolver o sistema')
                return
            else:
                k = matrix_triangular[j][i] / matrix_triangular[i][i]
                for l in range(len(matrix)):
                    matrix_triangular[j][l] = matrix_triangular[j][l] - k * matrix_triangular[i][l]
    return matrix_triangular
```

```python
matrix2 = np.array([[5, 2, 1], [2, 1, -1], [1, -1, -2]])
```

```python
triangular = gauss_triangular(matrix2)
```

```python
def resolve_system(matrix, vector):
    triangular = gauss_triangular(matrix)
    vector_triangular = np.zeros(len(matrix))
    for i in range(len(matrix)):
        vector_triangular[i] = vector[i]
        for j in range(i):
            vector_triangular[i] = vector_triangular[i] - triangular[i][j] * vector_triangular[j]
    return vector_triangular
```

```python
vector_triangular = resolve_system(matrix, vector)
```
