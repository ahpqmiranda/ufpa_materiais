# Stress Analysis

Stress analysis is a branch of engineering mechanics that involves the study of stresses and strains in materials and structures. It is used to determine the effects of external forces and loads on materials and structures, and to predict the behavior of these materials and structures under different loading conditions.

Stress analysis is commonly used in the design and testing of mechanical and structural components, such as aircraft wings, bridges, buildings, and automotive parts. It involves the use of mathematical models and computational tools to simulate the behavior of these components under different loading conditions, such as tension, compression, bending, and shear.

The results of stress analysis can be used to optimize the design of these components, to ensure that they can withstand the forces and loads that they are likely to experience in real-world conditions. Stress analysis can also be used to predict the failure of materials and structures, and to develop strategies for preventing or mitigating these failures.

Overall, stress analysis is an important tool for engineers and designers, as it enables them to design safer and more reliable structures and components.

### To calculate stress transformations on 2D situations, you can use the following equations:

σ_y = (σ_r + σ_t) / 2 - (σ_r - σ_t) / 2 * cos(2θ) - τ_rt * sin(2θ)

τ_xy = - (σ_r - σ_t) / 2 * sin(2θ) + τ_rt * cos(2θ)

where:

σ_x = stress in the x-direction
σ_y = stress in the y-direction
σ_r = radial stress (inward or outward) at a particular point
σ_t = tangential stress (circumferential) at a particular point
τ_rt = shear stress acting on a plane perpendicular to the radius and tangent to the circumference at a particular point
θ = angle (in degrees) between a line from the origin to the point of interest and the x-axis (measured counter-clockwise)

These equations are known as the stress transformation equations, and they can be used to convert the stresses acting on a plane at a particular point from one coordinate system to another. In 2D situations, there are two commonly used coordinate systems: the Cartesian coordinate system and the polar coordinate system.

To use these equations, you need to first determine the values of σ_r, σ_t, and τ_rt at the point of interest. These values can be obtained from experimental data or from stress analysis software. Once you have these values, you can substitute them into the stress transformation equations along with the angle θ to calculate the stresses in the x and y directions, as well as the shear stress τ_xy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the stresses acting on the plane
sigma_x = 50
sigma_y = -30
tau_xy = 20

# Calculate the principal stresses and the center of the Mohr circle
sigma_avg = (sigma_x + sigma_y) / 2
sigma_diff = (sigma_x - sigma_y) / 2
radius = np.sqrt(sigma_diff**2 + tau_xy**2)
sigma_1 = sigma_avg + radius
sigma_2 = sigma_avg - radius
center_x = sigma_avg
center_y = 0

# Calculate the projections of the stresses on the horizontal and vertical axes
sigma_x_proj = (sigma_x + sigma_y) / 2
sigma_y_proj = sigma_x_proj
tau_xy_proj = tau_xy

# Create an array of angles for plotting the Mohr circle
theta = np.linspace(0, 2*np.pi, 100)

# Calculate the x and y coordinates of the Mohr circle
circle_x = center_x + radius*np.cos(theta)
circle_y = center_y + radius*np.sin(theta)

# Create a new figure and axis for the Mohr circle plot
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the Mohr circle, the principal stresses, and the axis projections
ax.plot(circle_x, circle_y)
ax.plot([sigma_x, sigma_y], [tau_xy, -tau_xy], 'ro')
ax.plot([sigma_1, sigma_2], [0, 0], 'bo')
ax.plot([sigma_x_proj, sigma_x_proj], [-radius, radius], 'k--')
ax.plot([-radius, radius], [tau_xy_proj, tau_xy_proj], 'k--')

# Add labels to the plot
ax.set_xlabel('Normal stress (MPa)')
ax.set_ylabel('Shear stress (MPa)')
ax.set_title('Mohr circle')

# Set the axis limits to fit the plot
max_stress = max(abs(sigma_x), abs(sigma_y), abs(tau_xy))
ax.set_xlim([-max_stress*1.2, max_stress*1.2])
ax.set_ylim([-max_stress*1.2, max_stress*1.2])

# Display the plot
plt.show()
```
