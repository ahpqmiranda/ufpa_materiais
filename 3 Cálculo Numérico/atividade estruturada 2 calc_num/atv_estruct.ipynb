### Questão 1: converter o sistema em matriz triangular superior:
modo 1:

```python
# diagonalização de Gauss

def gauss_diagonal(matrix):
    for coluna in range(len(matrix)+1):
        for linha in range(coluna+1, len(matrix)):
            fator = matrix[linha][coluna]/matrix[coluna][coluna]
            for i in range(len(matrix[linha])):
                matrix[linha][i] = matrix[linha][i] - fator*matrix[coluna][i]
    return matrix
```

# Questão 2: Utilizar a triangulação de Gauss para resolver o sistema:

```python
def gauss(matrixcof, answers, x, tol):
    L = np.tril(matrixcof)
    U = matrixcof - L
    i = 0

    while True:
        x_init = deepcopy(x)
        x = np.dot(inv(L), answers - np.dot(U, x))
        i = i + 1

        if i > 2:
            if (norm(x - x_init)) / norm(x) < tol:
                return x, i
```

```python
matrix2 = np.array([[5, 2, 1], [2, 1, -1], [1, -1, -2]]) # matriz de coeficientes
results2 = np.array([9, 4, -1]) # vetor de respostas
x0 = np.array([2, 3, 4]) # vetor de inciação
u, v = gauss(matrix2, results2, x0, 0.005) # chamada da função
u = pd.Series(u) # transforma o vetor em um dataframe
texto = 'x1 = {:.3f} x2 = {:.3f} x3 = {:.3f} || n iterações: {}'.format(u[0], u[1], u[2], v) # texto com os valores de x1, x2 e x3 e a quantidade de iterações
texto
```

# Questão 3: Converter sistema de equações em matriz triangular superior e resolver:

```python
matrix_coef3 = [[7, 3, -2], [1, 4, -2], [2, -2, 6]] # matriz de coeficientes
results_3 = [15, 6, -4] # vetor de respostas
```

```python
def diagonaliza(matrix):
    for coluna in range(len(matrix)+1):
        for linha in range(coluna+1, len(matrix)):
            fator = matrix[linha][coluna]/matrix[coluna][coluna]
            for i in range(len(matrix[linha])):
                matrix[linha][i] = matrix[linha][i] - fator*matrix[coluna][i]
    return matrix
```

```python
matrix_triangular3 = diagonaliza(matrix_coef3)
```

```python
def solution_diagonal(matrix, answers):
    matrix = np.array(matrix)
    l = matrix.shape[0]
    D = np.linalg.det(matrix)

    det = np.zeros(l)

    for j in range(l):
        complement = deepcopy(matrix)
        for i in range(l):
            complement[i, j] = answers[i]
        det[j] = np.linalg.det(complement)

    answer = np.zeros(l)
    for i in range(len(det)):
        answer[i] = det[i] / D
    return answer
```

```python
solution_diagonal(matrix_triangular3, results_3)
```
