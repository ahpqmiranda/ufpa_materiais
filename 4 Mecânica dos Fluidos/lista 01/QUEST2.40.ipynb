```python
import sympy as sp
from IPython.display import Image
```

```python
sp.init_printing()
x, y, tau = sp.symbols('x y tau')
u = sp.Function('u')(y)
```

```python
Image(filename='quest2.40.png')
```

Dados Utilizados

```python
h = 1e-4   # [m] distncia entre as placas
mi = 1.14e-3  # [N.s/m] viscosidade dinmica da gua a 15C
umax = 0.10  # [m/s] velocidade mxima do escoamento
```

Hipteses simplificadoras
1. Fluido Newtoniano
2. Regime Permanente

Equao do perfil de velocidades do escoamento:

```python
r = (1 - (2*y/h) ** 2) * umax
r
```

Derivao do perfil de velocidades para encontrar o cisalhamento do fluido:

```python
dubydy = sp.simplify(sp.Derivative(r, y))
cis = sp.Eq(u.diff(y), dubydy)
cis
```

Substituio dos valores de cisalhamento e viscosidade para encontrar o valor das tenses viscosas do fluido:

```python
flu_newtoniano = sp.Eq(tau, mi * dubydy)
flu_newtoniano
```

O valor do cisalhamento do fluido :

```python
arrasto = flu_newtoniano.subs(y, h/2)
resultado = sp.solve(arrasto)
print('A tenso exercida pelo fluido na parede foi de {:.3f} N/m'.format(resultado[0]))
print('O perfil de velocidade caracterizado  o apresentado a seguir:')
```
