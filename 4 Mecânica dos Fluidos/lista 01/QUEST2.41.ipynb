```python
import numpy as np
import sympy as sp
from IPython.display import Image
```

```python
sp.init_printing()
x, y, tau, F = sp.symbols('x y tau F')
u = sp.Function('u')(y)
```

```python
Image(filename='quest2.41.png')
```

Dados Utilizados

```python
h = 1e-4            # [m] distância entre as placas
mi = 1.14e-3        # [N.s/m²] viscosidade dinâmica da água a 15ºC
umax = 5e-2         # [m/s] velocidade máxima do escoamento
area = 1            # [m²]
```

Hipóteses simplificadoras
1. Fluido Newtoniano
2. Regime Permanente

Derivação do perfil de velocidades para encontrar o cisalhamento do fluido:

```python
r = (1 - (2*y/h) ** 2) * umax
r
```

Derivação do perfil de velocidades para encontrar o cisalhamento do fluido:


```python
dubydy = sp.simplify(sp.Derivative(r, y))
cis = sp.Eq(u.diff(y), dubydy)
cis
```

Substituição dos valores de cisalhamento e viscosidade para encontrar o valor das tensões viscosas do fluido:

```python
flu_newtoniano = sp.Eq(tau, mi * dubydy)
flu_newtoniano
```

```python
arrasto = flu_newtoniano.subs(y, h/2)
resultado = sp.solve(arrasto)
print('A tensão exercida pelo fluido na parede foi de {:.3f} N/m²'.format(resultado[0]))
```

```python
force = sp.solve(sp.Eq(resultado[0], F / area))
print('A força atuante em uma área de 1 m² é de {:.3f} N'.format(force[0]))
```
