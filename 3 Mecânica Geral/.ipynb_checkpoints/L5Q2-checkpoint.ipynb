### Dados necess√°rios

```python
r = 8 # [ft]
Lg = 6 # [ft]
H = 10 # [ft]
pi = m.pi
```

A caixa d'√°gua √© composta por tr√™s sec√ß√µes, uma c√¥nica, uma cil√≠ndrica e outra esf√©rica, que ser√£o chamadas respectivamente de 1, 2 e 3, logo, temos os seguintes centroides das curvas em rela√ß√£o ao eixo Z e seus comprimentos:

```python
# Centroides das curvas
C1 = r / 2
C2 = r
C3 = 2 * r /pi

# Centroides das √°reas
CA1 = r / 3
CA2 = r / 2
CA3 = 4 * r / (pi * 3)

# Comprimentos dos arcos
L1 = m.dist([r,0],[0, Lg]) # dist√¢ncia euclidiana entre dois pontos
L2 = H
L3 = pi * r / 2

# √Åreas
A1 = r * Lg / 2
A2 = r * H
A3 = pi * r ** 2 / 4

# Tabela
data_cl = pd.DataFrame({
    'Centroides curvas':[C1, C2, C3],
    'Centroides √°reas':[CA1, CA2, CA3],
    'Comprimentos':[L1, L2, L3],
    '√Åreas': [A1, A2, A3]
})
data_cl
```

C√°lculo da √°rea superficial:
produto da coluna 'centroides C' com a 'Comprimentos'

```python
C = data_cl.iloc[0:,0:1].squeeze()
CA = data_cl.iloc[0:,1:2].squeeze()
L = data_cl.iloc[0:,2:3].squeeze()
A = data_cl.iloc[0:,3:4].squeeze()
```

√Årea de superf√≠cie do tanque;

```python
area = 2* pi * np.dot(C, L)
area =  0.092903 * area * units.m ** 2
area
```

Volume do tanque:

```python
volume =2 * pi * np.dot(CA, A)
volume = 0.0283168 * volume * units.m ** 3
volume
```
